### Q1: What is Matplotlib? Why is it used? Name five plots that can be plotted using the Pyplot module of
Matplotlib.

Matplotlib is a popular Python library used for creating high-quality, publication-ready visualizations. It provides a wide range of tools for constructing various types of graphs, plots, charts, and diagrams.

Matplotlib is used for a variety of purposes such as data exploration, presentation, and communication of data insights. It allows users to visualize data in a way that can be easily understood by others, making it an essential tool for data scientists, researchers, and analysts.

The Pyplot module of Matplotlib is one of the most commonly used modules, and it provides a simple interface for creating a variety of plots. Here are five types of plots that can be plotted using the Pyplot module of Matplotlib:

Line plot: A line plot is a simple plot where a series of data points are connected by a straight line. This type of plot is useful for visualizing trends over time or across different categories.

Bar plot: A bar plot is a plot where bars are used to represent the data points. It is useful for comparing data across different categories.

Scatter plot: A scatter plot is a plot where individual data points are plotted as dots on a graph. It is useful for visualizing the relationship between two variables.

Histogram: A histogram is a plot where the distribution of a single variable is shown. It is useful for visualizing the distribution of continuous data.

Box plot: A box plot is a plot where the distribution of a single variable is shown using quartiles. It is useful for comparing the distribution of data across different categories or groups.

### Q2: What is a scatter plot? Use the following code to generate data for x and y. Using this generated data
plot a scatter plot.

import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)

y = 3 + np.random.normal(0, 2, len(x))

Note: Also add title, xlabel, and ylabel to the plot.

A scatter plot is a type of data visualization that displays the relationship between two numerical variables. It consists of a set of points, where each point represents a pair of values, one for each variable.

To plot a scatter plot using the provided code, you can use the matplotlib library in Python. Here's the code:

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(3)

x = 3 + np.random.normal(0, 2, 50)
y = 3 + np.random.normal(0, 2, len(x))

plt.scatter(x, y)
plt.title('Scatter Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

The first line imports the necessary library, and the next two lines generate the data for x and y using the np.random.normal() function. The seed() function is used to ensure that the data is reproducible.

The plt.scatter() function is used to create the scatter plot, with x and y as inputs. The plt.title(), plt.xlabel(), and plt.ylabel() functions are used to add a title, x-axis label, and y-axis label to the plot, respectively.

The plt.show() function is used to display the plot. Running this code will produce a scatter plot with x on the x-axis and y on the y-axis, where each point represents a pair of values. The title of the plot is "Scatter Plot", and the x-axis label is "X" while the y-axis label is "Y".

### Q3: Why is the subplot() function used? Draw four line plots using the subplot() function.
Use the following data:

import numpy as np

For line 1: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([0, 100, 200, 300, 400, 500])

For line 2: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([50, 20, 40, 20, 60, 70])

For line 3: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([10, 20, 30, 40, 50, 60])

For line 4: x = np.array([0, 1, 2, 3, 4, 5]) and y = np.array([200, 350, 250, 550, 450, 150])

The subplot() function is used to create multiple plots within a single figure. It allows us to divide the figure into multiple subplots and specify which subplot we want to draw a particular plot in. This function takes three arguments: the number of rows, the number of columns, and the index of the current subplot.

Here's the code to draw four line plots using the subplot() function with the provided data:

```python
import numpy as np
import matplotlib.pyplot as plt

# First line plot
x1 = np.array([0, 1, 2, 3, 4, 5])
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)
plt.plot(x1, y1)
plt.title('Line 1')

# Second line plot
x2 = np.array([0, 1, 2, 3, 4, 5])
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)
plt.plot(x2, y2)
plt.title('Line 2')

# Third line plot
x3 = np.array([0, 1, 2, 3, 4, 5])
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)
plt.plot(x3, y3)
plt.title('Line 3')

# Fourth line plot
x4 = np.array([0, 1, 2, 3, 4, 5])
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)
plt.plot(x4, y4)
```

In this code, we first import the necessary libraries. Then, we define the data for each line plot using the provided arrays. We then use the subplot() function to create a 2x2 grid of subplots, and specify the index of each subplot for each line plot. We also add a title to each subplot using the title() function.

Finally, we use the show() function to display the figure with all four line plots. This code will produce a figure with four line plots, where each plot is contained in a separate subplot.

### Q4: What is a bar plot? Why is it used? Using the following data plot a bar plot and a horizontal bar plot.
import numpy as np

company = np.array(["Apple", "Microsoft", "Google", "AMD"])

profit = np.array([3000, 8000, 1000, 10000])

A bar plot is a type of plot used to represent data with rectangular bars. The length or height of each bar corresponds to the magnitude of the data being represented. Bar plots are commonly used to show comparisons between different categories or groups of data.

A bar plot can be used to compare the profit of different companies, as shown in the following code:

```python
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Vertical bar plot
plt.bar(company, profit)
plt.title("Profit of Companies")
plt.xlabel("Company")
plt.ylabel("Profit (in millions)")

plt.show()

# Horizontal bar plot
plt.barh(company, profit)
plt.title("Profit of Companies")
plt.xlabel("Profit (in millions)")
plt.ylabel("Company")

plt.show()
```

In this code, we first import the necessary libraries and define the data for the plot using the provided arrays. We then use the bar() function to create a vertical bar plot of the data, where the x-axis represents the companies and the y-axis represents the profit in millions. We also add a title, x-label, and y-label to the plot using the title(), xlabel(), and ylabel() functions, respectively.

Next, we use the barh() function to create a horizontal bar plot of the same data, where the y-axis represents the companies and the x-axis represents the profit in millions. We also add the same title, x-label, and y-label to this plot.

The resulting plots show the profit of each company using rectangular bars, with the length or height of each bar representing the magnitude of the data. The vertical bar plot shows the data in a traditional way, while the horizontal bar plot is useful when there are long category names and we want to save space.
